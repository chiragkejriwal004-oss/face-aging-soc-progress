## What is PyTorch?

PyTorch is an open-source deep learning framework developed by Meta.

It is widely used for:
- Computer Vision
- Natural Language Processing
- Deep Learning Research
- Neural Network Development

PyTorch uses tensors as its primary data structure.

```python
import torch

tensor = torch.tensor([1, 2, 3, 4, 5])

print("Tensor:")
print(tensor)
```

```python
import torch

a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

print("Addition:", a + b)
print("Multiplication:", a * b)
```

## Tensor Shapes

Every tensor has a shape that describes its dimensions.

Examples:
- Scalar -> Shape []
- Vector -> Shape [n]
- Matrix -> Shape [rows, columns]

```python
import torch

matrix = torch.tensor([[1, 2],
                       [3, 4]])

print("Matrix:")
print(matrix)

print("Shape:", matrix.shape)
```

## Automatic Differentiation

PyTorch provides autograd which automatically computes gradients.

This feature is used during neural network training and backpropagation.

```python
import torch

x = torch.tensor(2.0, requires_grad=True)

y = x**2

y.backward()

print("Gradient:", x.grad)
```
