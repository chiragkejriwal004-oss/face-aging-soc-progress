```python
import numpy as np

utkface_age_samples = np.random.randint(
    1,
    80,
    size=20
)

print("Sample Ages:")
print(utkface_age_samples)

print("Average Age:", np.mean(utkface_age_samples))
```

```python
import matplotlib.pyplot as plt

plt.hist(
    utkface_age_samples,
    bins=10
)

plt.title("Sample Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

sample_face = np.random.randint(
    0,
    255,
    (128,128,3),
    dtype=np.uint8
)

plt.imshow(sample_face)

plt.title("Sample Face Image")
plt.show()
```

```python
from PIL import Image

img = Image.fromarray(sample_face)

print("Image Size:", img.size)
```

```python
resized = img.resize((64,64))

print("Resized Shape:", resized.size)

resized
```

```python
import numpy as np

img_array = np.array(resized)

normalized = img_array / 255.0

print("Min:", normalized.min())
print("Max:", normalized.max())
```

```python
import torch

tensor_img = torch.tensor(
    normalized,
    dtype=torch.float32
)

print(tensor_img.shape)
```

```python
from torchvision import transforms

transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.ToTensor()
])

print(transform)
```
