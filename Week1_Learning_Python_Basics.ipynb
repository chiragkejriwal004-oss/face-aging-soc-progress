# Week 1 Learning - Python Basics

This notebook contains my practice work while learning Python fundamentals as part of the Summer of Code Face Aging Project.

Topics Covered:
- Variables and Data Types
- Conditional Statements
- Loops
- Functions
- Lists and Dictionaries
- NumPy Basics

```python
# Variables and Data Types

name = "Chirag"
age = 18
height = 5.9
student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", student)
```

```python
# Conditional Statements

num = 10

if num > 0:
    print("Positive Number")
elif num == 0:
    print("Zero")
else:
    print("Negative Number")
```

```python
# Loops

print("For Loop")

for i in range(5):
    print(i)

print("While Loop")

count = 0

while count < 5:
    print(count)
    count += 1
```

```python
# Functions

def square(x):
    return x * x

print(square(5))
print(square(10))
```

```python
# Lists

numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print(num)
```

```python
# Dictionaries

student = {
    "name": "Chirag",
    "roll": "25B0665",
    "branch": "Civil Engineering"
}

for key, value in student.items():
    print(key, ":", value)
```

```python
# NumPy Basics

import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
```
