```python
## What is TensorFlow?

TensorFlow is an open-source machine learning and deep learning framework developed by Google.

It is widely used for:
- Image Classification
- Object Detection
- Natural Language Processing
- Deep Learning Applications

TensorFlow provides efficient tools for building and training neural networks.
```

```python
import tensorflow as tf

tensor = tf.constant([1, 2, 3, 4, 5])

print("Tensor:")
print(tensor)
```

```python
import tensorflow as tf

a = tf.constant([1, 2, 3])
b = tf.constant([4, 5, 6])

print("Addition:", tf.add(a, b))
print("Multiplication:", tf.multiply(a, b))
```

## Tensor Shapes

TensorFlow stores data using tensors.

Examples:
- Scalar
- Vector
- Matrix
- Higher-dimensional tensors

Tensor shapes help define how data is organized.

```python
import tensorflow as tf

matrix = tf.constant([[1, 2],
                      [3, 4]])

print(matrix)
print("Shape:", matrix.shape)
```
