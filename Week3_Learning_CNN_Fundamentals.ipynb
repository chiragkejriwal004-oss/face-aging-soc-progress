```python
import numpy as np

image = np.array([
    [1, 2, 3, 0],
    [4, 5, 6, 1],
    [7, 8, 9, 2],
    [3, 4, 5, 6]
])

print(image)
```

```python
kernel = np.array([
    [1, 0],
    [0, -1]
])

print(kernel)
```

```python
def convolution(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    output = np.zeros((h-kh+1, w-kw+1))

    for i in range(h-kh+1):
        for j in range(w-kw+1):
            region = image[i:i+kh, j:j+kw]
            output[i,j] = np.sum(region * kernel)

    return output

feature_map = convolution(image, kernel)

print(feature_map)
```

```python
def max_pooling(feature_map, size=2):

    h, w = feature_map.shape

    output = np.zeros((h//size, w//size))

    for i in range(0, h, size):
        for j in range(0, w, size):

            region = feature_map[i:i+size, j:j+size]

            output[i//size, j//size] = np.max(region)

    return output

pooled = max_pooling(feature_map)

print(pooled)
```

```python
import torch.nn as nn

conv = nn.Conv2d(
    in_channels=1,
    out_channels=4,
    kernel_size=3
)

print(conv)
```

```python
relu = nn.ReLU()

sample = torch.tensor([
    [-2., -1., 0., 1., 2.]
])

print(relu(sample))
```

```python
pool = nn.MaxPool2d(2)

x = torch.rand(1,1,4,4)

print("Input Shape:", x.shape)

output = pool(x)

print("Output Shape:", output.shape)
```

```python
cnn = nn.Sequential(
    nn.Conv2d(1,16,3),
    nn.ReLU(),
    nn.MaxPool2d(2),
    nn.Conv2d(16,32,3),
    nn.ReLU()
)

print(cnn)
```
