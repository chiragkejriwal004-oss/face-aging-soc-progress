## What is a Neural Network?

A Neural Network is a machine learning model inspired by the human brain.

It consists of interconnected neurons organized into layers:
1. Input Layer
2. Hidden Layer(s)
3. Output Layer

Neural networks learn patterns from data and are widely used in image recognition, speech processing, and artificial intelligence.

```python
import numpy as np

inputs = np.array([1, 2, 3])
weights = np.array([0.5, 0.3, 0.2])

output = np.dot(inputs, weights)

print("Neuron Output:", output)
```

```python
## Activation Functions

Activation functions help neural networks learn complex patterns.

Common activation functions:
- ReLU
- Sigmoid
- Tanh

ReLU is the most commonly used activation function in deep learning.
```

```python
import numpy as np

x = np.array([-2, -1, 0, 1, 2])

relu = np.maximum(0, x)

print("Input:", x)
print("ReLU Output:", relu)
```

## Forward Propagation

Forward propagation is the process where:
- Inputs enter the network
- Calculations are performed
- Predictions are generated

This is the first step in training a neural network.

```python
import numpy as np

inputs = np.array([2, 4])

weights = np.array([0.6, 0.8])

bias = 1

output = np.dot(inputs, weights) + bias

print("Output after adding bias:", output)
```
